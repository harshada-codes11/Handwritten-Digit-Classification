# 🔢 MNIST Handwritten Digit Classification using Neural Networks

**Project Overview:**  
In this project, we build a deep learning model using a Neural Network (Multi-Layer Perceptron) to classify handwritten digits from the famous MNIST dataset. The model learns to recognize digits 0–9 from 28×28 grayscale images.

**Key Steps:**
1. Import libraries
2. Load and explore the MNIST dataset
3. Preprocess the data (normalize pixel values)
4. Build and compile the Neural Network
5. Train the model
6. Evaluate performance on test data
7. Visualize the confusion matrix
8. Build an interactive predictive system



## 📦 Step 1: Importing the Required Libraries

We import NumPy for numerical operations, Matplotlib and Seaborn for visualization, OpenCV for image processing, PIL for image display in Colab, and TensorFlow/Keras for building the neural network.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import cv2
from google.colab.patches import cv2_imshow
from PIL import Image

import tensorflow as tf
tf.random.set_seed(3)

from tensorflow import keras
from keras.datasets import mnist
from tensorflow.math import confusion_matrix

print(" All libraries imported successfully!")
print(f"TensorFlow version: {tf.__version__}")
```

## 📂 Step 2: Loading the MNIST Dataset

The MNIST dataset is built into Keras. It contains:
- **60,000 training images** (used to teach the model)
- **10,000 test images** (used to evaluate performance)
- Each image is **28×28 pixels**, grayscale (1 channel)
- Labels are integers from **0 to 9**

```python
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

print("Dataset loaded successfully!")
print(f"  Training images shape : {X_train.shape}  → 60,000 images of size 28×28")
print(f"  Training labels shape : {Y_train.shape}")
print(f"  Testing images shape  : {X_test.shape}   → 10,000 images of size 28×28")
print(f"  Testing labels shape  : {Y_test.shape}")
```

## 🔍 Step 3: Exploring the Data

Before training, it's important to understand the structure and content of our data. Let's visualize a few sample images and inspect their pixel values.

```python
print(f"Data type of X_train: {type(X_train)}")

print(f"Unique classes in Training data : {np.unique(Y_train)}")
print(f"Unique classes in Testing data  : {np.unique(Y_test)}")
```

```python
fig, axes = plt.subplots(3, 3, figsize=(8, 8))
fig.suptitle('Sample MNIST Training Images', fontsize=16, fontweight='bold')

sample_indices = [0, 1, 2, 10, 15, 25, 100, 200, 500]
for ax, idx in zip(axes.flatten(), sample_indices):
    ax.imshow(X_train[idx], cmap='gray')
    ax.set_title(f'Label: {Y_train[idx]}', fontsize=12)
    ax.axis('off')

plt.tight_layout()
plt.show()
print("📌 Each image is 28×28 pixels — grayscale (values: 0=black, 255=white)")
```

## ⚙️ Step 4: Data Preprocessing — Normalizing Pixel Values

**Why normalize?**  
Raw pixel values range from 0 to 255. Neural networks train faster and more stably when input values are small (between 0 and 1). We divide every pixel value by 255 to achieve this.

> All image dimensions in this dataset are already 28×28, so resizing is not needed.

```python
X_train = X_train / 255.0
X_test  = X_test  / 255.0

print(f"After normalization:")
print(f"  Min pixel value in training set: {X_train.min():.1f}")
print(f"  Max pixel value in training set: {X_train.max():.1f}")
print("Normalization complete — all values are now between 0 and 1")
```

## 🧠 Step 5: Building the Neural Network

**Architecture:**
- **Input Layer (Flatten):** Converts 28×28 image → flat vector of 784 values
- **Hidden Layer 1 (Dense, 50 neurons, ReLU):** Learns low-level features
- **Hidden Layer 2 (Dense, 50 neurons, ReLU):** Learns higher-level patterns
- **Output Layer (Dense, 10 neurons, Sigmoid):** One neuron per digit (0–9)

**Why ReLU?** It introduces non-linearity and avoids the vanishing gradient problem.  
**Why Sigmoid for output?** Outputs values between 0 and 1, interpreted as probabilities.

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),

    keras.layers.Dense(50, activation='relu'),

    keras.layers.Dense(50, activation='relu'),

    keras.layers.Dense(10, activation='sigmoid')
])

model.summary()
```

## 🔧 Step 6: Compiling the Model

Before training, we configure the model with:
- **Optimizer (Adam):** Adaptive learning rate — efficient and widely used
- **Loss Function (Sparse Categorical Crossentropy):** Used when labels are integers (0–9)
- **Metric (Accuracy):** Measures percentage of correct predictions

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("✅ Model compiled successfully!")
print("   Optimizer : Adam")
print("   Loss      : Sparse Categorical Crossentropy")
print("   Metric    : Accuracy")
```

## 🚀 Step 7: Training the Model

We train for **10 epochs** — one epoch means the model has seen all 60,000 training images once. Watch how accuracy improves and loss decreases with each epoch!

```python
history = model.fit(
    X_train,
    Y_train,
    epochs=10
)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Model Training History', fontsize=15, fontweight='bold')

ax1.plot(history.history['accuracy'], 'b-o', linewidth=2, markersize=6)
ax1.set_title('Training Accuracy per Epoch', fontsize=13)
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.set_ylim([0.8, 1.0])
ax1.grid(True, alpha=0.3)
ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))

ax2.plot(history.history['loss'], 'r-o', linewidth=2, markersize=6)
ax2.set_title('Training Loss per Epoch', fontsize=13)
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 📊 Step 8: Evaluating the Model on Test Data

We evaluate the trained model on 10,000 test images it has **never seen before**. This gives us an unbiased estimate of real-world performance.

```python
test_loss, test_accuracy = model.evaluate(X_test, Y_test, verbose=1)

print(f"\n{'='*40}")
print(f"  Training Accuracy : ~99.2%")
print(f"  Test Accuracy     : {test_accuracy * 100:.2f}%")
print(f"  Test Loss         : {test_loss:.4f}")
print(f"{'='*40}")
```

## 🔮 Step 9: Making Predictions

`model.predict()` returns the **probability** of each class (0–9) for every image. We use `np.argmax()` to pick the class with the highest probability as our final predicted label.

```python
Y_pred = model.predict(X_test)

print(f"Shape of predictions: {Y_pred.shape}")
print(f"\nProbabilities for first test image (true label = {Y_test[0]}):")
print(Y_pred[0])
print(f"\nThe model is most confident about digit: {np.argmax(Y_pred[0])}")
```

```python
Y_pred_labels = [np.argmax(pred) for pred in Y_pred]

plt.figure(figsize=(3, 3))
plt.imshow(X_test[0], cmap='gray')
plt.title(f'True Label: {Y_test[0]} | Predicted: {Y_pred_labels[0]}', fontsize=12)
plt.axis('off')
plt.show()
```

## 📉 Step 10: Confusion Matrix

The confusion matrix shows us:
- **Diagonal cells** = correctly classified digits ✅
- **Off-diagonal cells** = misclassifications ❌

This helps us identify which digits the model confuses with each other most often.

```python
conf_mat = confusion_matrix(Y_test, Y_pred_labels)

plt.figure(figsize=(12, 8))
sns.heatmap(
    conf_mat,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=range(10),
    yticklabels=range(10)
)
plt.title('Confusion Matrix — MNIST Test Set', fontsize=15, fontweight='bold', pad=15)
plt.ylabel('True Labels', fontsize=13)
plt.xlabel('Predicted Labels', fontsize=13)
plt.tight_layout()
plt.show()

print("\nRaw confusion matrix:")
print(conf_mat.numpy())
```

## 🖼️ Step 11: Predictive System — Classify Your Own Image

This is the most exciting part! You can upload any image of a handwritten digit and the model will predict which digit it is.

**How it works:**
1. Read the image with OpenCV
2. Convert to grayscale (model was trained on grayscale)
3. Resize to 28×28 pixels (model input size)
4. Normalize pixel values (divide by 255)
5. Reshape to match model input format
6. Predict using the trained model

```python
input_image_path = '/content/sample_data/3-digit.PNG'

input_image = cv2.imread(input_image_path)

print("Original image:")
cv2_imshow(input_image)
print(f"Original shape: {input_image.shape}")
```

```python
grayscale = cv2.cvtColor(input_image, cv2.COLOR_RGB2GRAY)
print(f"After grayscale conversion: {grayscale.shape}")

input_image_resize = cv2.resize(grayscale, (28, 28))
print(f"After resize to 28×28: {input_image_resize.shape}")

print("Resized 28×28 image:")
cv2_imshow(input_image_resize)

input_image_resize = input_image_resize / 255.0

image_reshaped = np.reshape(input_image_resize, [1, 28, 28])
print(f"Final input shape for model: {image_reshaped.shape}")
```

```python
input_prediction = model.predict(image_reshaped)
input_pred_label = np.argmax(input_prediction)

print(f"\nPrediction probabilities: {input_prediction[0].round(4)}")
print(f"\n🎯 The Handwritten Digit is recognised as: {input_pred_label}")
```

## 🤖 Step 12: Full Interactive Predictive System

Run the cell below and enter the path of **any handwritten digit image** when prompted. The system will display the image and tell you which digit it detected.

```python
def predict_digit(image_path):
    """
    Predicts the digit in a given image using the trained neural network.

    Parameters:
        image_path (str): Path to the input image file

    Returns:
        int: The predicted digit (0–9)
    """
    img = cv2.imread(image_path)

    if img is None:
        print(" Error: Could not read the image. Please check the file path.")
        return None

    print(" Input Image:")
    cv2_imshow(img)

    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    resized = cv2.resize(gray, (28, 28))
    normalized = resized / 255.0
    model_input = np.reshape(normalized, [1, 28, 28])

    # Predict
    prediction_probs = model.predict(model_input, verbose=0)
    predicted_label = np.argmax(prediction_probs)

    print(f"\n The Handwritten Digit is recognised as: {predicted_label}")
    return predicted_label


image_path = input('📂 Enter path of the image to be predicted: ')
predict_digit(image_path)
```
